```python
# Dependencies and setup.
import os
import csv
import pandas as pd


# Import ethnicity data.
path_eth1 = os.path.join("ethnicity", "Enrollment Report_Statewide_Districts_Ethnicity_2014-2015.csv")
path_eth2 = os.path.join("ethnicity", "Enrollment Report_Statewide_Districts_Ethnicity_2015-2016.csv")
path_eth3 = os.path.join("ethnicity", "Enrollment Report_Statewide_Districts_Ethnicity_2016-2017.csv")
path_eth4 = os.path.join("ethnicity", "Enrollment Report_Statewide_Districts_Ethnicity_2017-2018.csv")
path_eth5 = os.path.join("ethnicity", "Enrollment Report_Statewide_Districts_Ethnicity_2018-2019.csv")

# Convert ethnicity data to pandas data frame.
df_eth1 = pd.read_csv(path_eth1)
df_eth2 = pd.read_csv(path_eth2)
df_eth3 = pd.read_csv(path_eth3)
df_eth4 = pd.read_csv(path_eth4)
df_eth5 = pd.read_csv(path_eth5)
```

```python
# Change 2018-2019 Travis County ethnicity data column name font to Title Case to match other years.

df_eth5.columns = map(str.title, df_eth5.columns)
df_eth5
```

```python
# Change 2018-2019 Travis County ethnicity data column names to match other years.

df_eth5.rename(columns = {"Year":"Year", "Region":"Region", "County Name":"County Name", "District":"District Number",
                          "District Name":"District Name", "Ethnicity":"Ethnicity Name", "Charter Status":"Charter Status", 
                          "Enrollment":"Enrollment by Ethnicity" }, inplace = True) 
df_eth5
```

```python
# Swap 2018-2019 Travis County ethnicity data column order (for District Number and District Name) to match other years.

cols1 = list(df_eth5.columns)
a, b = cols1.index('District Number'), cols1.index('District Name')
cols1[a], cols1[b] = cols1[b], cols1[a]
df_eth5 = df_eth5[cols1]

cols2 = list(df_eth5.columns)
c, d = cols2.index('Ethnicity Name'), cols2.index('Charter Status')
cols2[c], cols2[d] = cols2[d], cols2[c]
df_eth5_clean = df_eth5[cols2]

df_eth5_clean
```
