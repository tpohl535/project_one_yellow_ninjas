```python
#dependencies set up
import pandas as pd
```

```python
#import cleaned CSVs
path_1 = "Resources/aisd_meal_reimbursement_years_15_19_UPDATED.csv"
path_2 = "Resources/schools_geolocation_aisd.csv"
```

```python
#open and read paths
meal_reimb_df = pd.read_csv(path_1)
aisd_geolocation = pd.read_csv(path_2)
```

```python
#merge datasets on SiteName
merged_df = pd.merge(meal_reimb_df, aisd_geolocation, on="SiteName", how="outer")
merged_df
```

```python
#output merged_df to csv
merged_df.to_csv("Resources/schools_reimbursement_location_merged.csv", index=False)
```
