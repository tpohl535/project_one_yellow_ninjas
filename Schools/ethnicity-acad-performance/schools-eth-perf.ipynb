```python
# Dependencies and setup.
import os
import csv
import pandas as pd

# Import ethnicity data.
path_eth15 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2014-2015.csv")
path_eth16 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2015-2016.csv")
path_eth17 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2016-2017.csv")
path_eth18 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2017-2018.csv")
path_eth19 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2018-2019.csv")

# Convert ethnicity data to pandas data frame.
df_eth15 = pd.read_csv(path_eth15)
df_eth16 = pd.read_csv(path_eth16)
df_eth17 = pd.read_csv(path_eth17)
df_eth18 = pd.read_csv(path_eth18)
df_eth19 = pd.read_csv(path_eth19)
```

```python
# Display first dataframe to understand data structure.
df_eth15
```

```python
# Filter each dataframes to display only Travis County ethnicity data.

df_eth15_travis = df_eth15.loc[df_eth15["County Name"] == "TRAVIS COUNTY"]
df_eth15_travis = df_eth15_travis.drop(df_eth15_travis.columns[1], axis=1)
df_eth15_travis
```

```python
df_eth16_travis = df_eth16.loc[df_eth16["County Name"] == "TRAVIS COUNTY"]
df_eth16_travis = df_eth16_travis.drop(df_eth16_travis.columns[1], axis=1)
df_eth16_travis
```

```python
df_eth17_travis = df_eth17.loc[df_eth17["County Name"] == "TRAVIS COUNTY"]
df_eth17_travis = df_eth17_travis.drop(df_eth17_travis.columns[[1, 7]], axis=1)
df_eth17_travis
```

```python
df_eth18_travis = df_eth18.loc[df_eth18["County Name"] == "TRAVIS COUNTY"]
df_eth18_travis = df_eth18_travis.drop(df_eth18_travis.columns[[1, 7]], axis=1)
df_eth18_travis
```

```python
# Displaying 5th dataframe for 2018-2019 ethnicity data reveals differences compared to other years in column order, column names and fonts.

df_eth19_travis = df_eth19.loc[df_eth19["COUNTY NAME"] == "TRAVIS COUNTY"]
df_eth19_travis 
```

```python
# Change 2018-2019 Travis County ethnicity data column name font to match other years.

df_eth19.columns = map(str.title, df_eth19.columns)
df_eth19
```

```python
# Change 2018-2019 Travis County ethnicity data column names to match other years.

df_eth19.rename(columns = {"Year":"Year", "Region":"Region", "County Name":"County Name", "District":"District Number",
                          "District Name":"District Name", "Campus":"Campus Number", "Campus Name":"Campus Name", "Ethnicity":"Ethnicity Name",
                           "Enrollment":"Enrollment by Ethnicity" }, inplace = True) 
df_eth19
```

```python
# Swap 2018-2019 Travis County ethnicity data column order (for District Number and District Name) to match other years.

cols1 = list(df_eth19.columns)
a, b = cols1.index("District Number"), cols1.index("District Name")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_eth19 = df_eth19[cols1]

cols2 = list(df_eth19.columns)
c, d = cols2.index("Campus Number"), cols2.index("Campus Name")
cols2[c], cols2[d] = cols2[d], cols2[c]
df_eth19_clean = df_eth19[cols2]
df_eth19_clean 
```

```python
# Filter 2018-2019 dataframe to display only Travis County ethnicity data.

df_eth19_clean_travis = df_eth19_clean.loc[df_eth19_clean["County Name"] == "TRAVIS COUNTY"]
df_eth19_clean_travis = df_eth19_clean_travis.drop(df_eth19_clean_travis.columns[[1, 8]], axis=1)
df_eth19_clean_travis
```

```python
# Concatenate all 5 years' dataframes for Travis County ethnicity data, and reset index to avoid duplicate indices.

df_eth5yrs_travis = pd.concat([df_eth15_travis, df_eth16_travis, df_eth17_travis, df_eth18_travis, df_eth19_clean_travis])
df_eth5yrs_travis=df_eth5yrs_travis.reset_index(drop=True)
df_eth5yrs_travis
```

```python
# Output to a CSV file the final concatenated dataframe for 5 years of Travis County ethnicity data.

df_eth5yrs_travis.to_csv("Output_ethnicity_5yrs_Travis.csv", index=False)
```
