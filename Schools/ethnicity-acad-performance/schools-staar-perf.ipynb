```python
# ACADEMIC PERFORMANCE DATA (STAAR SCORES)

# Dependencies and setup.
import os
import csv
import pandas as pd

# Import STAAR scores by campus in Travis County.
path_staar15 = os.path.join("STAAR_Travis_Campuses_2014-2015.csv")
path_staar16 = os.path.join("STAAR_Travis_Campuses_2015-2016.csv")
path_staar17 = os.path.join("STAAR_Travis_Campuses_2016-2017.csv")
path_staar18 = os.path.join("STAAR_Travis_Campuses_2017-2018.csv")
path_staar19 = os.path.join("STAAR_Travis_Campuses_2018-2019.csv")

# Convert STAAR data to pandas data frame.
df_staar15 = pd.read_csv(path_staar15)
df_staar16 = pd.read_csv(path_staar16)
df_staar17 = pd.read_csv(path_staar17)
df_staar18 = pd.read_csv(path_staar18)
df_staar19 = pd.read_csv(path_staar19)

# Import STAAR score definitions.
path_def15 = os.path.join("STAAR_definitions_2014-2015.csv")
path_def16 = os.path.join("STAAR_definitions_2015-2016.csv")
path_def17 = os.path.join("STAAR_definitions_2016-2017.csv")
path_def18 = os.path.join("STAAR_definitions_2017-2018.csv")
path_def19 = os.path.join("STAAR_definitions_2018-2019.csv")

# Convert STAAR score definitions to pandas data frame.
df_def15 = pd.read_csv(path_def15)
df_def16 = pd.read_csv(path_def16)
df_def17 = pd.read_csv(path_def17)
df_def18 = pd.read_csv(path_def18)
df_def19 = pd.read_csv(path_def19)

# Import ethnicity data output to get campus names in Travis County.
path_campus = os.path.join("Output_ethnicity_5yrs_clean_Travis.csv")
# Convert STAAR data to pandas data frame.
df_campus = pd.read_csv(path_campus)
```

```python
# Format CAMPUS column data and store in new column Campus_Number for all 5 years.

df_staar15["Campus_Number"] = df_staar15["CAMPUS"].str.rsplit("'").str[-1] 
# Swap CAMPUS and Campus_Number columns.
cols1 = list(df_staar15.columns)
a, b = cols1.index("CAMPUS"), cols1.index("Campus_Number")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_staar15 = df_staar15[cols1]

df_staar16["Campus_Number"] = df_staar16["CAMPUS"].str.rsplit("'").str[-1] 
cols1 = list(df_staar16.columns)
a, b = cols1.index("CAMPUS"), cols1.index("Campus_Number")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_staar16 = df_staar16[cols1]

df_staar17["Campus_Number"] = df_staar17["CAMPUS"].str.rsplit("'").str[-1] 
cols1 = list(df_staar17.columns)
a, b = cols1.index("CAMPUS"), cols1.index("Campus_Number")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_staar17 = df_staar17[cols1]

df_staar18["Campus_Number"] = df_staar18["CAMPUS"].str.rsplit("'").str[-1] 
cols1 = list(df_staar18.columns)
a, b = cols1.index("CAMPUS"), cols1.index("Campus_Number")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_staar18 = df_staar18[cols1]
df_staar18.head()

df_staar19["Campus_Number"] = df_staar19["CAMPUS"].str.rsplit("'").str[-1] 
cols1 = list(df_staar19.columns)
a, b = cols1.index("CAMPUS"), cols1.index("Campus_Number")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_staar19 = df_staar19[cols1]
df_staar19.head()
```

```python
# Drop old CAMPUS column for all 5 years.

df_staar15 = df_staar15.drop(df_staar15.columns[43], axis=1)
df_staar16 = df_staar16.drop(df_staar16.columns[43], axis=1)
df_staar17 = df_staar17.drop(df_staar17.columns[43], axis=1)
df_staar18 = df_staar18.drop(df_staar18.columns[43], axis=1)
df_staar19 = df_staar19.drop(df_staar19.columns[55], axis=1)

df_staar19.head()
```

```python
# Set "Campus_Number" as index column for all 5 years.

df_staar15 = df_staar15.set_index("Campus_Number")
df_staar16 = df_staar16.set_index("Campus_Number")
df_staar17 = df_staar17.set_index("Campus_Number")
df_staar18 = df_staar18.set_index("Campus_Number")
df_staar19 = df_staar19.set_index("Campus_Number")
```

```python
# Display STAAR data definitions for all 5 years, and drop extra columns where needed for consistency.

df_def15.head()

df_def16 = df_def16.drop(df_def16.columns[[1, 2]], axis=1)
df_def17 = df_def17.drop(df_def17.columns[[1, 2]], axis=1)
df_def18 = df_def18.drop(df_def18.columns[[1, 2]], axis=1)
df_def19 = df_def19.drop(df_def19.columns[[1, 2]], axis=1)

df_def19.head()
```

```python
# Edit definition string to keep only relevant information for all 5 years.

df_def15["Description"] = df_def15["Description"].str.split(',').str[-1]
df_def16["LABEL"] = df_def16["LABEL"].str.split(',').str[-1]
df_def17["LABEL"] = df_def17["LABEL"].str.split(',').str[-1]
df_def18["LABEL"] = df_def18["LABEL"].str.split(',').str[-1]
df_def19["LABEL"] = df_def19["LABEL"].str.split(',').str[-1]

df_def19.head()
```

```python
# Edit column names in STAAR dataframe based on element definitions provided in other dataframe for all 5 years.

df_staar15 = df_staar15.rename(columns=dict(df_def15.values))
df_staar16 = df_staar16.rename(columns=dict(df_def16.values))
df_staar17 = df_staar17.rename(columns=dict(df_def17.values))
df_staar18 = df_staar18.rename(columns=dict(df_def18.values))
df_staar19 = df_staar19.rename(columns=dict(df_def19.values))

df_staar19.head()
```

```python
# Filter and keep only columns with All Test scores for all 5 years.

df_staar15_filt = df_staar15.filter(regex='All Tests')
df_staar16_filt = df_staar16.filter(regex='All Tests')
df_staar17_filt = df_staar17.filter(regex='All Tests')
df_staar18_filt = df_staar18.filter(regex='All Tests')
df_staar19_filt = df_staar19.filter(regex='All Tests')

df_staar19_filt.head()
```

```python
# Drop columns not needed for all 5 years.

df_staar15_drop = df_staar15_filt.drop(df_staar15_filt.columns[[3, 6, 9, 10, 12]], axis=1)
df_staar15_clean = df_staar15_drop.reset_index()
df_staar15_clean["Campus_Number"] = df_staar15_clean["Campus_Number"].astype(int)

df_staar16_drop = df_staar16_filt.drop(df_staar16_filt.columns[[3, 6, 9, 10, 12]], axis=1)
df_staar16_clean = df_staar16_drop.reset_index()
df_staar16_clean["Campus_Number"] = df_staar16_clean["Campus_Number"].astype(int)

df_staar17_drop = df_staar17_filt.drop(df_staar17_filt.columns[[3, 6, 9, 10, 12]], axis=1)
df_staar17_clean = df_staar17_drop.reset_index()
df_staar17_clean["Campus_Number"] = df_staar17_clean["Campus_Number"].astype(int)

df_staar18_drop = df_staar18_filt.drop(df_staar18_filt.columns[[3, 6, 9, 10, 12]], axis=1)
df_staar18_clean = df_staar18_drop.reset_index()
df_staar18_clean["Campus_Number"] = df_staar18_clean["Campus_Number"].astype(int)

df_staar19_drop = df_staar19_filt.drop(df_staar19_filt.columns[[0, 4, 7, 10, 11, 13, 15, 16, 17]], axis=1)
df_staar19_clean = df_staar19_drop.reset_index()
df_staar19_clean["Campus_Number"] = df_staar19_clean["Campus_Number"].astype(int)
df_staar19_clean.head()
```

```python
# Update column names to math other data for consistency with ethnicity dataset for all 5 years.

df_staar15_clean.columns = ["Campus_Number", "Multiracial-AllSub Score", "Asian-AllSub Score","Native Hawaiian/Other Pacific Islander-AllSub Score", "Black or African American-AllSub Score", 
                         "Econ Disadv-AllSub Score", "Hispanic/Latino-AllSub Score", "American Indian or Alaska Native-AllSub Score", "At Risk-AllSub Score", "White-AllSub Score"]                     
df_staar16_clean.columns = ["Campus_Number", "Multiracial-AllSub Score", "Asian-AllSub Score","Native Hawaiian/Other Pacific Islander-AllSub Score", "Black or African American-AllSub Score", 
                         "Econ Disadv-AllSub Score", "Hispanic/Latino-AllSub Score", "American Indian or Alaska Native-AllSub Score", "At Risk-AllSub Score", "White-AllSub Score"]                     
df_staar17_clean.columns = ["Campus_Number", "Multiracial-AllSub Score", "Asian-AllSub Score","Native Hawaiian/Other Pacific Islander-AllSub Score", "Black or African American-AllSub Score", 
                         "Econ Disadv-AllSub Score", "Hispanic/Latino-AllSub Score", "American Indian or Alaska Native-AllSub Score", "At Risk-AllSub Score", "White-AllSub Score"]                     
df_staar18_clean.columns = ["Campus_Number", "Multiracial-AllSub Score", "Asian-AllSub Score","Native Hawaiian/Other Pacific Islander-AllSub Score", "Black or African American-AllSub Score", 
                         "Econ Disadv-AllSub Score", "Hispanic/Latino-AllSub Score", "American Indian or Alaska Native-AllSub Score", "At Risk-AllSub Score", "White-AllSub Score"]                     
df_staar19_clean.columns = ["Campus_Number", "Multiracial-AllSub Score", "Asian-AllSub Score","Native Hawaiian/Other Pacific Islander-AllSub Score", "Black or African American-AllSub Score", 
                         "Econ Disadv-AllSub Score", "Hispanic/Latino-AllSub Score", "American Indian or Alaska Native-AllSub Score", "At Risk-AllSub Score", "White-AllSub Score"]                     
df_staar19_clean.head()
```

```python
# Add Year column for year of school ending to each of the 5 years' datasets.

df_staar15_clean["Year"] ='2015'
df_staar16_clean["Year"] ='2016'
df_staar17_clean["Year"] ='2017'
df_staar18_clean["Year"] ='2018'
df_staar19_clean["Year"] ='2019'
df_staar19_clean
```

```python
# Reorder columns for all 5 years.

df_staar15_clean = df_staar15_clean[["Year", "Campus_Number", "Asian-AllSub Score", "Hispanic/Latino-AllSub Score", "Native Hawaiian/Other Pacific Islander-AllSub Score", 
                                    "Multiracial-AllSub Score", "White-AllSub Score", "Black or African American-AllSub Score", "American Indian or Alaska Native-AllSub Score", 
                                    "Econ Disadv-AllSub Score", "At Risk-AllSub Score"]]
df_staar16_clean = df_staar16_clean[["Year", "Campus_Number", "Asian-AllSub Score", "Hispanic/Latino-AllSub Score", "Native Hawaiian/Other Pacific Islander-AllSub Score", 
                                    "Multiracial-AllSub Score", "White-AllSub Score", "Black or African American-AllSub Score", "American Indian or Alaska Native-AllSub Score", 
                                    "Econ Disadv-AllSub Score", "At Risk-AllSub Score"]]
df_staar17_clean = df_staar17_clean[["Year", "Campus_Number", "Asian-AllSub Score", "Hispanic/Latino-AllSub Score", "Native Hawaiian/Other Pacific Islander-AllSub Score", 
                                    "Multiracial-AllSub Score", "White-AllSub Score", "Black or African American-AllSub Score", "American Indian or Alaska Native-AllSub Score", 
                                    "Econ Disadv-AllSub Score", "At Risk-AllSub Score"]]
df_staar18_clean = df_staar18_clean[["Year", "Campus_Number", "Asian-AllSub Score", "Hispanic/Latino-AllSub Score", "Native Hawaiian/Other Pacific Islander-AllSub Score", 
                                    "Multiracial-AllSub Score", "White-AllSub Score", "Black or African American-AllSub Score", "American Indian or Alaska Native-AllSub Score", 
                                    "Econ Disadv-AllSub Score", "At Risk-AllSub Score"]]
df_staar19_clean = df_staar19_clean[["Year", "Campus_Number", "Asian-AllSub Score", "Hispanic/Latino-AllSub Score", "Native Hawaiian/Other Pacific Islander-AllSub Score", 
                                    "Multiracial-AllSub Score", "White-AllSub Score", "Black or African American-AllSub Score", "American Indian or Alaska Native-AllSub Score", 
                                    "Econ Disadv-AllSub Score", "At Risk-AllSub Score"]]
df_staar19_clean.head()
```

```python
# Concatenate all 5 years' dataframes for Travis county schools STAAR results data, and reset index to avoid duplicate indices.

df_staar5yrs_travis = pd.concat([df_staar15_clean, df_staar16_clean, df_staar17_clean, df_staar18_clean, df_staar19_clean])
df_staar5yrs_clean_travis = df_staar5yrs_travis.reset_index(drop=True)
df_staar5yrs_clean_travis
```

```python
# Output to a CSV file the final concatenated dataframe for 5 years of Travis County schools STAAR data.

df_staar5yrs_clean_travis.to_csv("Output_STAAR_5yrs_clean_Travis.csv", index=False)
```
