```python
# ACADEMIC PERFORMANCE DATA (STAAR SCORES)

# Dependencies and setup.
import os
import csv
import pandas as pd

# Import STAAR scores by campus in Travis County.
path_staar15 = os.path.join("STAAR_Travis_Campuses_2014-2015.csv")
path_staar16 = os.path.join("STAAR_Travis_Campuses_2015-2016.csv")
path_staar17 = os.path.join("STAAR_Travis_Campuses_2016-2017.csv")
path_staar18 = os.path.join("STAAR_Travis_Campuses_2017-2018.csv")
path_staar19 = os.path.join("STAAR_Travis_Campuses_2018-2019.csv")

# Convert STAAR data to pandas data frame.
df_staar15 = pd.read_csv(path_staar15)
df_staar16 = pd.read_csv(path_staar16)
df_staar17 = pd.read_csv(path_staar17)
df_staar18 = pd.read_csv(path_staar18)
df_staar19 = pd.read_csv(path_staar19)

# Import STAAR score definitions.
path_def15 = os.path.join("STAAR_definitions_2014-2015.csv")
path_def16 = os.path.join("STAAR_definitions_2015-2016.csv")
path_def17 = os.path.join("STAAR_definitions_2016-2017.csv")
path_def18 = os.path.join("STAAR_definitions_2017-2018.csv")
path_def19 = os.path.join("STAAR_definitions_2018-2019.csv")

# Convert STAAR score definitions to pandas data frame.
df_def15 = pd.read_csv(path_def15)
df_def16 = pd.read_csv(path_def16)
df_def17 = pd.read_csv(path_def17)
df_def18 = pd.read_csv(path_def18)
df_def19 = pd.read_csv(path_def19)

# Import ethnicity data output to get campus names in Travis County.
path_campus = os.path.join("Output_ethnicity_5yrs_clean_Travis.csv")
# Convert STAAR data to pandas data frame.
df_campus = pd.read_csv(path_campus)
```

```python
# Format CAMPUS column data and store in new column Campus_Number.
df_staar15["Campus_Number"] = df_staar15["CAMPUS"].str.rsplit("'").str[-1] 

# Swap CAMPUS and Campus_Number columns.
cols1 = list(df_staar15.columns)
a, b = cols1.index("CAMPUS"), cols1.index("Campus_Number")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_staar15 = df_staar15[cols1]

df_staar15.head()
```

```python
# # Drop old CAMPUS column
df_staar15 = df_staar15.drop(df_staar15.columns[43], axis=1)
df_staar15.head()
```

```python
#  Set "Campus_Number" as index column.
df_staar15 = df_staar15.set_index("Campus_Number")
```
