```python
permit_df = pd.read_csv("cleaned_again_permitdata.csv")
permit_df
```

```python
# Issue-1 how many different types of permits are allowed yearly in only EAST AUSTIN
types_df = permit_df.groupby(
    ["calendar_year_issued", "original_city", "original_zip", "permit_type_desc"]
)["permit_type_desc"].value_counts()
types_df
```

```python
permit_df.shape  # dimensionality of the DataFrame
```

```python
permit_df["permit_class_mapped"].value_counts()
```

```python
# Issue-2 Display the number of residential and commercial permits that are finalized in a year
permitclass_df = permit_df.groupby(
    ["calendar_year_issued", "permit_class_mapped", "status_current"]
)["permit_class_mapped"].value_counts()
permitclass_df
```

```python
# Issue-3 Which zipcode in East Austin has the highest permits
count_zip = (
    permit_df.groupby(["original_zip", "original_city", "permit_type_desc"])[
        "permit_type_desc"
    ]
    .value_counts()
    .sort_values(ascending=False)
    .index[0]
)
count_zip
```

```python
# Issue-4 Which zipcode in East Austin has the lowest permits
count_zip = (
    permit_df.groupby(["original_zip", "original_city", "permit_type_desc"])[
        "permit_type_desc"
    ]
    .value_counts()
    .sort_values(ascending=True)
    .index[0]
)
count_zip
```
